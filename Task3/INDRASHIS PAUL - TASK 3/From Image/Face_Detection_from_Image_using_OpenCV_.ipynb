## Importing required Libraries 

```
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

## Face detection
I'll be using the `detectMultiscale` module of the classifier. This function will return a rectangle with coordinates(x,y,w,h) around the detected face.

```
def detect_faces(cascade, img, scaleFactor=1.3):
    # Loading the image and Creating a copy of the image to prevent any changes to the original one.
    img_raw = image_copy = cv2.imread(img)

    # Converting the test image to gray scale as opencv face detector expects gray images
    gray_image = cv2.cvtColor(image_copy, cv2.COLOR_BGR2GRAY)

    # Applying the haar classifier to detect faces
    faces_rect = cascade.detectMultiScale(gray_image, scaleFactor = scaleFactor, minNeighbors=5)

    # Drawing a Green Rectangle according to the coordinates provided by the HAAR classifier over the detected faces
    for (x, y, w, h) in faces_rect:
        cv2.rectangle(image_copy, (x, y), (x+w, y+h), (0, 0, 255), 3)

    # Printing the number of faces in the image
    print("The number of faces found: ", len(faces_rect))

    # Printing the detected image
    cv2_imshow(image_copy) # This is used because cv2.imshow() causes the Notebook to crash
```

### Loading the classifier for frontal face

```
cascade = cv2.CascadeClassifier('haarcascade_frontalface_alt2.xml')
```

## Detecting Faces

```
detect_faces(cascade, '/content/pexels-photo-614810.jpeg')
```
